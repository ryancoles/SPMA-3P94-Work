```python
%reset -f 
import json
with open('season_team_stats.json', 'r') as f:
     teams_dict = json.load(f)
```

```python

ANA_dict = teams_dict['ANA']
ANA_dict
```

```python
ANA_gp = ANA_dict['GamesPlayed']
ANA_wins = ANA_dict['Wins']
print(ANA_gp, ANA_wins)
```

```python
ANA_wp = ANA_wins / ANA_gp
ANA_wp = round(ANA_wp, 4)
ANA_wp
```

```python
ANA_GF2 = ANA_dict['GoalsFor'] **2
ANA_GA2 = ANA_dict['GoalsAllowed'] **2
print(ANA_GF2, ANA_GA2)
```

```python

ANA_EWP = ANA_GF2 / (ANA_GF2 + ANA_GA2)
ANA_EWP = round(ANA_EWP, 4)
ANA_EWP
```

```python

work_dict = teams_dict.copy()
work_dict
```

```python

for team in work_dict:
    print ('Start of loop for ', team)
    WINS = work_dict[team]['Wins']
    GP = work_dict[team]['GamesPlayed']
    WP = WINS / GP
    work_dict[team]['WinPercentage'] = round(WP, 2)
    print('End of loop for ', team)
print('Completed loop')
```

```python
for team in work_dict:
#     print('Start of loop for ', team)
    GF2 = work_dict[team]['GoalsFor']**2
    work_dict[team]['GoalsFor2'] = GF2
#     print('End of loop for ', team)
print('Completed loop')
```

```python

for team in work_dict:
#     print('Start of loop for ', team)
    GA2 = work_dict[team]['GoalsAllowed']**2
    work_dict[team]['GoalsAllowed2'] = GA2
#     print('End of loop for ', team)
print('Completed loop')
```

```python

for team in work_dict:
#     print('Start of loop for ', team)
    EWP = work_dict[team]['GoalsFor2'] / (work_dict[team]['GoalsFor2'] + work_dict[team]['GoalsAllowed2'])
    work_dict[team]['EWP'] = round(EWP, 4)
#     print('End of loop for ', team)
print('Completed loop')
```
